## Question 4.1

Show that an approximation to the Wald statistic can be developed from the second-order Taylor expansion of the log-likelihood as follows. For this problem, focus on just one of the regression parameters, say $_j$.

1. Write the first three terms of the Taylor series expansion of $(\beta_j; y)$ expanded about $\hat{}_j$.

2. Rearrange to show that the Wald statistic is approximately equal to $2\{(\beta_j; y) - (\hat{\beta}_j; y) \}$, and hence show that the Wald statistic is approximately equivalent to a likelihood ratio test when $_j  \hat{}_j$ is small.

### Answers

#### Answer to 4.1, 1

***Write the first three terms of the Taylor series expansion of $(\beta_j; y)$ expanded about $\hat{}_j$.***

##### The log likelihood

Let's start by expressing the log likelihood .

Recall that the likelihood of a given parameter value $\theta$ given the data $\vec{X}$ is the probability of the data given parameters:

$$(\theta | \vec{X}) = p(\vec{X}|\theta) = \prod_{i=1}^n p(x_i | \theta)$$

We take the logarithm of this to get the log likelihood:

$$ = \text{ln}((\theta | \vec{X})) = \text{ln}(p(\vec{X}|\theta)) = \sum_{i=1}^n \text{ln}(p(x_i | \theta))$$

##### The Taylor Series

If you're not familiar with Taylor series, you might want to check out [Khan Academy's lesson](https://www.khanacademy.org/math/ap-calculus-bc/bc-series-new/bc-10-11/v/maclaurin-and-taylor-series-intuition).  Taylor series are a way of expressing any smooth, continuous function with a polynomial series.  Generically, a Taylor series looks like:

$f(x) = \sum_{n=0}^\infty \frac{f^{(n)}(k)}{n!} (x-k)^n$

> $= \frac{f^{(0)}(k)}{0!} (x-k)^0 + \frac{f^{(1)}(k)}{1!} (x-k)^1 + \frac{f^{(2)}(k)}{2!} (x-k)^2 + \frac{f^{(3)}(k)}{3!} (x-k)^3 + ...$
> $= f(k) + f'(k)(x-k) + \frac{f''(a)}{2}(x-k)^2 + \frac{f'''(k)}{3}(x-k)^3 + ...$

where $k$ is the value of $x$ for which we are evaluating $f(x)$, $f(x=k)$.  

Of course, we don't generally want to find an infinity of terms, so we so just find a few and accept the truncated expression as an approximation of $f(x)$ like

$f(x) \approx f(k) + f'(k)(x-k) + \frac{f''(k)}{2} (x-k)^2$

In the context of our question, we have

$ \approx (\hat{\beta}_j) + \frac{\partial}{\partial \beta_j}\left[(\hat{\beta}_j)  \right] (\beta_j - \hat{\beta}_j)+ \frac{1}{2}\frac{\partial^2}{\
partial \beta_j^2} \left[(\hat{\beta}) \right] (\beta_j - \hat{\beta}_j)^2$

It turns out that $\frac{\partial}{\partial \beta_j}\left[(\hat{\beta}_j)  \right] = 0$ such that we're left with

$ \approx (\hat{\beta}_j) + \frac{1}{2}\frac{\partial^2}{\
partial \beta_j^2} \left[(\hat{\beta}) \right] (\beta_j - \hat{\beta}_j)^2$

#### Answer to 4.1, 2

***Rearrange to show that the Wald statistic is approximately equal to $2\{(\beta_j; y) - (\hat{\beta}_j; y) \}$, and hence show that the Wald statistic is approximately equivalent to a likelihood ratio test when $_j  \hat{}_j$ is small.***

##### The log-likelihood ratio

In the question, $2\{(\beta_j^0; y) - (\hat{\beta}_j; y)\}$ is the a logged ratio of 2 likelihoods, scaled by $2$: one for the null-hypothesis value of $\beta_j$ and one for the estimated value of $\beta_j$.

$$\begin{aligned}
\text{ln}\left(\frac{(\beta_j^0 | y)}{(\hat{\beta}_j | y)}\right) = (\beta_j^0 | y) - (\hat{\beta}_j | y)
\end{aligned}$$

This gets scaled by $2$ because this logged ratio is the numerator of a fraction with a denominator of $2$; multiplying by $2$ turns it into a Chi-square distribution (exact for normal distributions, approximate otherwise), which is convenient for hypothesis testing.

$$2\{(\beta_j^0; y) - (\hat{\beta}_j; y)\} \sim ^2_1$$

##### The Wald Statistic

Here, the Wald statistic is

$$\begin{aligned}
W &= \frac{(\hat{\beta}_j - \beta_j^0)^2}{\hat{\text{var}}[\hat{\beta}_j]} \\
&= \frac{(\hat{\beta}_j - \beta_j^0)^2}{1/\mathcal{I}(\hat{\beta}_j)} \\
&= (\hat{\beta}_j - \beta_j^0)^2 \times \mathcal{I}(\hat{\beta}_j) \\
&= (\hat{\beta}_j - \beta_j^0)^2 \times -\left[\frac{\partial^2}{\partial \beta_j^2} \left[(\hat{\beta}) \right] \right]
\end{aligned}$$

##### Comparison between log-likelihood ratio and Wald Statistic

When we last left the log-likelihood expansion, we had

$ \approx (\hat{\beta}_j) + \frac{1}{2}\frac{^2}{\partial \beta_j^2} \left[(\hat{\beta}) \right] (\beta_j - \hat{\beta}_j)^2$

Let us rearrange this to get

$2\{ - (\hat{\beta})\} \approx \frac{^2}{\partial \beta_j^2} \left[(\hat{\beta}) \right] (\beta_j - \hat{\beta}_j)^2$

Now recall that

$\mathcal{I}(\hat{\beta}_j) = -\left[\frac{\partial^2}{\partial \beta_j^2} \left[(\hat{\beta}) \right] \right] = \frac{1}{\text{var}(\hat{\beta}_j)}$

Plugging this in, we get that

$2\{ - (\hat{\beta})\} \approx \frac{1}{\text{var}} (\beta_j - \hat{\beta}_j)^2 = W$

### Answer to 4.3, 1

***Working with just one observation, use the definition of the expected value to show that***

Alright.  Well, a generic definition for expected value is

$$[a|b] = \int_{-\infty}^{\infty} a \times p(a|b) \ da$$

where $a$ and $b$ can be data or parameters.  If $a$ is not conditioned on anything, then this simplifies to

$$[a] = \int_{-\infty}^{\infty} a \times p(a) \ da$$

It's also probably helpful to know that

$$U(\zeta) = \frac{d}{d \zeta} \left[(y|\zeta) \right]$$

where $(y|\zeta)$ is the log likelihood.  If $\zeta$ contains >1 parameter, then we use partial derivatives instead.

$$U(\zeta) = \frac{\partial}{\partial \zeta} \left[(y|\zeta) \right]$$

So now we have

$$\begin{align}
[U(\zeta)] &= \int_{-\infty}^{\infty} U(\zeta) \times p(\zeta|y) \ dy \\
&= \int_{-\infty}^{\infty} \frac{\partial}{\partial \zeta} \left[(y|\zeta) \right] \times p(\zeta|y) \ dy \\
&= \int_{-\infty}^{\infty} \frac{\partial}{\partial \zeta} \left[\text{ln}(p(\zeta|y)) \right] \times p(\zeta|y) \ dy \\
&= \int_{-\infty}^{\infty} \left(\frac{1}{p(\zeta|y} \frac{\partial}{\partial \zeta}[p(\zeta|y)] \right) \times p(\zeta|y) \ dy \\

\end{align}$$

which is the first thing we're meant to show.

***Using that $[U(\zeta)] = 0$ and the definition of the variance, show that $\text{var}[U(\zeta)] = [U(\zeta)U(\zeta)^T]$, which is $\mathcal{I}(\zeta)$ (assuming the order of the integration and differentiation can be reversed).***

$$\begin{align}

\end{align}$$

&= \frac{n\hat{\sigma^2} \left((\hat{\sigma^2}-1) - \text{ln}[\hat{\sigma^2}] \right)}{\hat{\sigma^2}} \\
&= n\left((\hat{\sigma^2}-1) - \text{ln}[\hat{\sigma^2}] \right) \\
&= n\left(\hat{\sigma^2} - 1 - \text{ln}[\hat{\sigma^2}] \right) \\
\end{align}$$

### Answer to 4.4, 9
***Show that $W = S = L$ in this example.***

We've already shown that

$$W(\mu_0=0), S(\mu_0=0), L(\mu_0=0) = \frac{n \hat{\mu}}{\sigma^2}$$

So this is done for $\mu$.  What about $\sigma^2$?

$$W(\sigma^2_0=1) = \frac{n(\hat{\sigma^2}-1)^2}{2 (\sigma^2)^2}$$
$$S(\sigma^2_0=1) = \frac{\left(\sum_{i=1}^n (y_i - \mu)^2 - n \right)^2}{2n}$$
$$L(\sigma^2_0=1) = n\left(\hat{\sigma^2} - 1 - \text{ln}[\hat{\sigma^2}] \right)$$

These are not equal expressions, at least for most values of $\hat{\sigma^2}$.  However, they are asymtoptically equal as $n \rightarrow \infty$.  We can this easily for $W$ and $L$--imagine $n=\infty$ and they both become $\infty$--and we can see it for $S$ if we rearrange a bit like

$$\begin{align}
S(\sigma^2_0=1) &= \frac{\left(\sum_{i=1}^n (y_i - \mu)^2 - n \right)^2}{2n} \\
&= \frac{(n\hat{\sigma^2} - n)^2}{2n} \\
&= \frac{(n(1-\hat{\sigma^2}))^2}{2n} \\
&= \frac{n^2(1-\hat{\sigma^2})^2}{2n} \\
&= \frac{n(1-\hat{\sigma^2})^2}{2} \\
\end{align}$$

***Find the score function.***

$$\begin{align}
U(\mu) &= \frac{d}{d \mu}[] \\
&= \frac{d}{d\mu}\left[-\left(n \ \text{ln}(\mu) + \frac{1}{\mu}\sum_{i=1}^n y_i \right) \right] \\
